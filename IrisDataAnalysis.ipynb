```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

data = load_iris()
d = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
dfx = pd.DataFrame(data=data.data, columns=d)
print("{} is X data frame".format(dfx))
dfy = pd.DataFrame(data=data.target, columns=['Species'])
print("{} is Y dataframe".format(dfy))
```

[Problem 6] Creating a diagram

It is important to visualize and confirm the data. Please do the following three things. In addition to Matplotlib, seaborn is often used as a plotting library.

- Make a pie chart of the number of samples per label (displayed as a percentage)
- Select features one by one and use a boxplot to visualize the distribution of data for each label
- Select features one by one and visualize the distribution of data for each label using a violin diagram

Boxplots and violin plots display similar plots. Please explain how they are different and what are the advantages of each.

```python
import matplotlib.pyplot as plt

#.loc is based data selecting method which means that is name of row and columns.
#iloc is based indexed data selecting method which means that is index of specific row and columns
# in this case i used .iloc. I think this is more optimal.

plt.pie(df.iloc[:,0])
plt.title("pie chart")
plt.show() 

plt.boxplot(df.iloc[:,1])
plt.title("box plot")
plt.show() 

plt.violinplot(df.iloc[:,2])
plt.title("violin plot")
plt.show() 
     
```
