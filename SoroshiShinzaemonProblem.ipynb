[Question 1] Number of rice grains on the 100th day
Write a code that calculates how many grains of rice you received in total on the 100th day, and output the answer. Also, please use a line graph to show the transition of "the number of rice received on that day" and "the total number of rice received by that day" from the 1st day to the 100th day. (The number of rice on the vertical axis, the number of days on the horizontal axis)

```python
import matplotlib.pyplot as plt
%matplotlib inline

number_of_rice = 1
number_of_grains_of_rice = []
total_number_of_grains_of_rice = []
total_grains_of_rice = 0

for i in range(100):

  total_grains_of_rice = total_grains_of_rice + number_of_rice
  number_of_grains_of_rice.append(number_of_rice)
  number_of_rice = number_of_rice * 2
  total_number_of_grains_of_rice.append(total_grains_of_rice)
  
print("The total number of grains of rice on the day from 1 to {} is".format(len(total_number_of_grains_of_rice)))
print("{}".format(total_number_of_grains_of_rice[-1]))
print("The number of grains of rice on the {}th day is".format(len(total_number_of_grains_of_rice)))
print("{}".format(number_of_grains_of_rice[-1]))

plt.title("The number of rice grains")
plt.xlabel("The number of days")
plt.ylabel("The number of rice")
plt.plot(number_of_grains_of_rice)
plt.show()

plt.title("The total number of rice grains you receive")
plt.xlabel("The number of days")
plt.ylabel("The number of rice")
plt.plot(total_number_of_grains_of_rice)
plt.show()
```

[Question 2] Number of rice grains other than the 100th day
In this anecdote, the number of days is ambiguous, such as 30 days or 81 days. Let's create a function with the number of days as an argument so that it can handle any story . After creating a function, please answer how many grains of rice you will receive for the number of days you take the course. Please output a line graph as well as Problem 1.

The template below contains not only the framework as a function, but also comments about arguments and return values. Please refer to it and rewrite the comment to make it easier to understand.


```python
def compute_sorori_shinzaemon(n_days):
#A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

  total_grains_of_rice = 0

  number_of_grains_of_rice = []
  total_number_of_grains_of_rice = []

  for i in range(n_days):
    number_of_grains_of_rice.append(2**i)
    total_grains_of_rice = total_grains_of_rice + number_of_grains_of_rice[i]
    total_number_of_grains_of_rice.append(total_grains_of_rice)

  return total_number_of_grains_of_rice, number_of_grains_of_rice

n_days=100
total_number_of_grains_of_rice, number_of_grains_of_rice = compute_sorori_shinzaemon(n_days)

#The line graph to represent the change in the number of rice grains you receive on that day
plt.title("The number of rice grains")
plt.xlabel("The number of days")
plt.ylabel("The number of rice")
plt.plot(number_of_grains_of_rice)
plt.show()

#The line graph to represent the total number of rice grains you receive by that day 
plt.title("The total number of rice grains you receive")
plt.xlabel("The number of days")
plt.ylabel("The number of rice")
plt.plot(total_number_of_grains_of_rice, color = "red")
plt.show()
```
