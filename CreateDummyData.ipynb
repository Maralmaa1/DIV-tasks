[Question 1] Create random numbers
Generate 500 random numbers from a two-dimensional normal distribution with mean (-3, 0) and covariance matrix given below.

```python
import numpy as np
import matplotlib.pyplot as plt

mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]] 
np.random.seed(0)
x, y = np.random.multivariate_normal(mean, cov, 500).T
a = x,y
plt.plot(x, y, ".", alpha=0.5)
plt.axis('equal')
plt.show()
```
